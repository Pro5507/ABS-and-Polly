```python
import requests
from PIL import image
from io import BytesIO
from config import HF_API_KEY

API_URL = "https://api-inference.huggingface.co/models/stabilityai/stable-diffusion-3-medium-diffusers"
def generate_image_from_text(prompt: str)->Image.Image:
  headers = {"Authorization": f"Bearer {HF_API_KEY}"}
  payload = {"inputs": prompt}
  try:
    response = requests.post(API_URL, headers=headers, json=payload, timeout=30)
    response.raise_for_status()
    if 'Image' in response.headers.get('Content-Type',''):
      image = Image.open(BytesIO(response.content))
      return image
    else:
      rais Exception("The response is not an image. It might be an error message.")
  except requests.exceptions.RequestException as e:
    raise Exception(f"Request failed: {e}")
def main():
  print("Welcome to the Text-to-Image generator")
  print("Type 'exit' to quit.")
  while True:
    prompt = input("Enter a description for the impage you want to generate:\n").strip()
    if prompt.lower=="exit":
      print("Good Bye!")
      break
    print("\nGenerating image.....\n")
    try:
      image = generate_image_from_text(prompt)
      image.show()
      save_option = input("Do you want to save this image? (yes/no):").strip().lower()
      if save_option == "yes":
        file_name = input("Enter a name for th image file (without extension):").strip() or "generated_image"
```
