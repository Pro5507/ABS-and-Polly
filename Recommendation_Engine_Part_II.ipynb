```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from math import sqrt
```

```python
ratings_df = pd.read_csv('ratings.csv')
ratings_df.head()
```

```python
movie_df = pd.read_csv('movies.csv')
movie_df.head()
```

```python
movie_df['year'] = movie_df.title.str.extract('(\(\d\d\d\d\))', expand=True)
movie_df.head()
```

```python
movie_df['year'] = movie_df.year.str.extract('(\d\d\d\d)', expand=True)
movie_df.head()
```

```python
movie_df['title'] = movie_df.title.str.extract('(\d\d\d\d)', expand=True)
movie_df.head()
```

```python
movie_df['genres'] = movie_df.genres.str.split('|')
movie_df.head()
```

```python
movie_copy = movie_df.copy()

for index, row in movie_df.iterrows():
  for genre in row['genres']:
    movie_copy.at[index, genre] = 1
movie_copy.head()
```
