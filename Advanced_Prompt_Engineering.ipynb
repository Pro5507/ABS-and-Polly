```python
import os
import time
from google import genai
from google.genai import types
import config

def generate_response(prompt, temprature=0.5):
  try:
    client = genai.Client(api_key=config.GEMINI_API_KEY)
    contents = [
        types.Content(role="user", parts=[types.Part.from_text(text=prompt),],),
    ]
    generate_content_config = types.GenerateContentConfig(temprature=temprature, response_mime_type="text/plain",)
    response = client.models.generate_content(model="gemini-2.0-flash", contents=contents, config=generate_content_config,)
    return response.text

 except Exception as e:
   return f"Error generating response: {str(e)}"

def temprature_prompt_activity():
  print("=" * 80)
  print("ADVANCED PROMPT ENGINEERING: TEMPRATURE & INSTRUCTION-BASED PROMPTS")
  print("=" * 80)
  print("\nIn this activity, we'll explore: ")
  print("1. How temprature affects AI creativity and randomness")
  print("2. How instruction-based prompt can control AI outputs")

  print("\n" + "-" * 40)
  print("PART 1: TEMPRATURE EXPLORATION")
  print("-" * 40)

  base_prompt = input("\nEnter a creative prompt (e.g., 'Write a story about a robot learning to paint'): ")
```
